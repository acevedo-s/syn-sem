```python
import jax.numpy as jnp
import numpy as np

A = np.load("/home/acevedo/syn-sem/results/txt_var_syn/modelA_llama/modelB_llama/match_var_matching/n_files_21/min_token_length_7/similarities/random_centers_1/Nbits_0/n_tokens_7/avg_tokens_1/batch_shuffle_0/layer_A_8/layer_B_8/centers_A.npy")
B = np.load("/home/acevedo/syn-sem/results/txt_var_syn/modelA_llama/modelB_llama/match_var_matching/n_files_21/min_token_length_7/similarities/random_centers_1/Nbits_0/n_tokens_7/avg_tokens_1/batch_shuffle_0/layer_A_8/layer_B_8/centers_B.npy")
A.shape

import matplotlib.pyplot as plt
A = A / jnp.linalg.norm(A, axis=1, keepdims=True)
B = B / jnp.linalg.norm(B, axis=1, keepdims=True)

result = jnp.tensordot(A, B, axes=[1, 1])
plt.hist(result.flatten(), bins=50)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Result')
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarities between all rows of A and B
cos_sim = cosine_similarity(A, B)

# Plot histogram
plt.hist(cos_sim.flatten(), bins=50)
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.title('Histogram of Cosine Similarities between rows of A and B')
plt.show()
```
