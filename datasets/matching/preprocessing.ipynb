```python
data = []
with open("matching_template_pairs.txt", "r") as f:
    for line in f:
        parts = line.strip().split('\t')
        if len(parts) == 3:
            sentence1, sentence2, tags = parts
            data.append((sentence1, sentence2, tags))

print(data[-1])
```

```python
# Open files for writing
with open("sentences1.txt", "w") as f1, \
     open("sentences2.txt", "w") as f2, \
     open("tags.txt", "w") as f3:
    
    for sentence1, sentence2, tags in data:
        f1.write(sentence1 + "\n")
        f2.write(sentence2 + "\n")
        f3.write(tags + "\n")
```
