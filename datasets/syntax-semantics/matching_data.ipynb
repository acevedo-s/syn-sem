```python
from pathlib import Path
from collections import defaultdict

src_path = Path("semantics/original_sentences_semantics.txt")
tgt_path = Path("syntax-twins/original_sentences_syntax.txt")

def norm(s: str) -> str:
    return " ".join(s.strip().split())

# <index> gives, for each sentence in tgt_path, the line where it appears
index = defaultdict(list)
with tgt_path.open("r", encoding="utf-8") as f:
    for i, line in enumerate(f):  # 0-based
        key = norm(line)
        if key:
            index[key].append(i)

# positions[k] gives the line number in tgt_path of a line in src_path, using index
positions = []
with src_path.open("r", encoding="utf-8") as f:
    for line in f:
        key = norm(line)
        positions.append(index.get(key, []))  # [] if missing
```

```python
### Number of shared sentences between src and tgt paths
len(positions) - [1 for p in positions if len(p) == 0].__len__()
```
