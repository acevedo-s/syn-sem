```python
from pathlib import Path
from collections import defaultdict

sem_path = Path("semantics/original_sentences_semantics.txt")
syn_path = Path("syntax-twins/original_sentences_syntax.txt")
pos_path = Path("syntax-twins/POS_templates.txt")

def norm(s: str) -> str:
    return " ".join(s.strip().split())

# <index> gives, for each sentence in syn_path, the line where it appears
index = defaultdict(list)
syn_sentences = []
with syn_path.open("r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        key = norm(line)
        index[key].append(i)
        syn_sentences.append(key)

# positions_syn[k] gives the line number in syn_path of a line in sem_path, using index
positions_syn = []
positions_sem = []
sem_sentences = []
with sem_path.open("r", encoding="utf-8") as f:
    for line_idx, line in enumerate(f):
        key = norm(line)
        sem_sentences.append(key)
        if key in index:
          positions_syn.append(index.get(key)[0])
          positions_sem.append(line_idx)
```

```python
shared_sentences = [sem_sentences[p] for p in positions_sem]
print(shared_sentences.__len__())
```
