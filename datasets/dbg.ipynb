```python
# File paths
sem_file = "/home/acevedo/syn-sem/datasets/txt/sem/second/matching/english/all_candidate_paraphrase_pairs.txt"
syn_file = "/home/acevedo/syn-sem/datasets/txt/syn/second/matching/syntax_matched_sentences.txt"

# Step 1: Load third-column values from the semantic file
with open(sem_file, "r", encoding="utf-8") as f:
    sem_lines = [line.strip().split("\t") for line in f if line.strip()]
    sem_targets = [line[2] for line in sem_lines if len(line) >= 3]

# Step 2: Load first and second column values from the syntax file
with open(syn_file, "r", encoding="utf-8") as f:
    syn_lines = [line.strip().split("\t") for line in f if line.strip()]
    syn_sents = set()
    for line in syn_lines:
        if len(line) >= 2:
            syn_sents.add(line[0])
            syn_sents.add(line[1])

# Step 3: Check which target sentences are missing from the syntax file
missing = [sent for sent in sem_targets if sent not in syn_sents]

# Step 4: Report results
if missing:
    print(f"{len(missing)} sentences from column 3 of the semantic file are missing:")
    for m in missing:
        print(m)
else:
    print("All third-column sentences in the semantic file appear in the syntax file.")
```

```python
# File paths
syn_path = "/home/acevedo/syn-sem/datasets/txt/syn/second/matching/sentences0.txt"
sem_path_0 = "/home/acevedo/syn-sem/datasets/txt/sem/second/matching/english/sentences0.txt"
sem_path_1 = "/home/acevedo/syn-sem/datasets/txt/sem/second/matching/english/sentences1.txt"

# Load sentences from syntax file
with open(syn_path, "r", encoding="utf-8") as f:
    syn_sentences = set(line.strip() for line in f if line.strip())

# Load sentences from both semantic files
with open(sem_path_0, "r", encoding="utf-8") as f:
    sem_sentences_0 = set(line.strip() for line in f if line.strip())

with open(sem_path_1, "r", encoding="utf-8") as f:
    sem_sentences_1 = set(line.strip() for line in f if line.strip())

# Union of semantic sentences
sem_sentences_all = sem_sentences_0 | sem_sentences_1

# Check for missing sentences
missing = syn_sentences - sem_sentences_all

# Report results
if missing:
    print(f"{len(missing)} sentence(s) from the syntax file are missing in both semantic files:")
    for m in missing:
        print(m)
else:
    print("All sentences from the syntax file are present in at least one semantic file.")
```
