```python
data = []
with open("deduplicated_mismatched_examples.txt", "r") as f:
    for line in f:
        parts = line.strip().split('\t')
        if len(parts) == 5:
            sentence1, sentence2, tags1, tags2, length = parts
            data.append((sentence1, sentence2, tags1, tags2, length))
print(data[-1])
```

```python
# Open files for writing
with open("sentences0.txt", "w") as f1, \
     open("sentences1.txt", "w") as f2, \
     open("tags0.txt", "w") as f3, \
     open("tags1.txt", "w") as f4, \
     open("lengths.txt", "w") as f5:
    
    for sentence1, sentence2, tags1, tags2, length in data:
        f1.write(sentence1 + "\n")
        f2.write(sentence2 + "\n")
        f3.write(tags1 + "\n")
        f4.write(tags2 + "\n")
        f5.write(length + "\n")
```

```python
from collections import Counter

# Extract lengths and convert to integers if needed
lengths = [int(length) for _, _, _, _, length in data]

# Count frequency of each length
length_counts = Counter(lengths)

# Print sorted distribution
for length, count in sorted(length_counts.items()):
    print(f"Length {length}: {count} examples")
```
