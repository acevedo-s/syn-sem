```python
from collections import defaultdict

path0    = './'
filename = 'deduplicated_matched_examples.txt'

# 1. Read your data
data = []
with open(f"{path0}{filename}", "r") as f:
    for line in f:
        s1, s2, tags = line.strip().split('\t')
        data.append((s1, s2, tags))

# 2. Build tag→ID mapping
unique_tags = sorted({tags for _, _, tags in data})
tag_to_id = {tags: idx for idx, tags in enumerate(unique_tags, start=0)}

# 3. Open four files and write out
with open(f"{path0}sentences0.txt", "w") as f1, \
     open(f"{path0}sentences1.txt", "w") as f2, \
     open(f"{path0}tags.txt",        "w") as f3, \
     open(f"{path0}group_ids.txt",   "w") as f4:

    for s1, s2, tags in data:
        gid = tag_to_id[tags]
        f1.write(s1   + "\n")
        f2.write(s2   + "\n")
        f3.write(tags + "\n")
        f4.write(str(gid) + "\n")
```
