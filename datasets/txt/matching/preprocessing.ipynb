```python
path0 = './'
filename = 'deduplicated_matched_examples.txt'
# path0 = './low_overlap/'
# filename = 'low_overlap_deduplicated_matched_examples.txt'
data = []
with open(f"{path0}{filename}", "r") as f:
    for line in f:
        parts = line.strip().split('\t')
        if len(parts) == 3:
            sentence1, sentence2, tags = parts
            data.append((sentence1, sentence2, tags))

print(data[-1])
print(len(data))
```

```python
# Open files for writing
with open(f"{path0}sentences0.txt", "w") as f1, \
     open(f"{path0}sentences1.txt", "w") as f2, \
     open(f"{path0}tags.txt", "w") as f3:
    
    for sentence1, sentence2, tags in data:
        f1.write(sentence1 + "\n")
        f2.write(sentence2 + "\n")
        f3.write(tags + "\n")
```
