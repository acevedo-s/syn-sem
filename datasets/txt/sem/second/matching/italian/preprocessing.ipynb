```python
# Concatenate the files and split columns into two output files
input_files = ['translations_1.txt', 'translations_2.txt', 'translations_3.txt']

with open('sentences0.txt', 'w', encoding='utf-8') as out0, open('sentences1.txt', 'w', encoding='utf-8') as out1:
  for fname in input_files:
    with open(fname, 'r', encoding='utf-8') as f:
      for line in f:
        parts = line.rstrip('\n').split('\t')
        if len(parts) >= 2:
          out0.write(parts[0] + '\n')
          out1.write(parts[1] + '\n')
```
