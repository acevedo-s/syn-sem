```python
with open('translations.txt', 'r', encoding='utf-8') as trans_file, \
  open('/home/acevedo/syn-sem/datasets/txt/sem/second/matching/english/sentences0.txt', 'r', encoding='utf-8') as s0_file, \
  open('/home/acevedo/syn-sem/datasets/txt/sem/second/matching/english/sentences1.txt', 'r', encoding='utf-8') as s1_file, \
  open('aligned_filtered_translations.txt', 'w', encoding='utf-8') as out_file:

    # Build a mapping from English sentence to translation line
    trans_map = {}
    for trans_line in trans_file:
        eng = trans_line.split('\t')[0].strip()
        trans_map[eng] = trans_line

    s0_file.seek(0)
    s1_file.seek(0)

    for s0_line, s1_line in zip(s0_file, s1_file):
        s0_eng = s0_line.strip()
        s1_eng = s1_line.strip()
        # Write the translation if present, in the same order as s0/s1
        if s0_eng in trans_map:
            out_file.write(trans_map[s0_eng])
        elif s1_eng in trans_map:
            out_file.write(trans_map[s1_eng])
```

```python
with open('aligned_filtered_translations.txt', 'r', encoding='utf-8') as infile, \
  open('sentences0.txt', 'w', encoding='utf-8') as out0, \
  open('sentences1.txt', 'w', encoding='utf-8') as out1:
    for line in infile:
     parts = line.rstrip('\n').split('\t')
     if len(parts) >= 2:
      out0.write(parts[0] + '\n')
      out1.write(parts[1] + '\n')
```

```python
with open('/home/acevedo/syn-sem/datasets/txt/sem/second/matching/english/all_candidate_paraphrase_pairs.txt', 'r', encoding='utf-8') as cand_file:
  # Build a set of all candidate sentences (from both columns)
  candidate_sentences = set()
  for line in cand_file:
    parts = line.rstrip('\n').split('\t')
    if len(parts) >= 1:
      candidate_sentences.add(parts[0])
    if len(parts) >= 2:
      candidate_sentences.add(parts[1])

with open('translations.txt', 'r', encoding='utf-8') as trans_file:
  missing_sentences = []
  for line in trans_file:
    eng_sentence = line.split('\t')[0].strip()
    if eng_sentence not in candidate_sentences:
      missing_sentences.append(eng_sentence)

print(f"Number of missing sentences: {len(missing_sentences)}")
if missing_sentences:
  print("Missing sentences:")
  for s in missing_sentences:
    print(s)
```
