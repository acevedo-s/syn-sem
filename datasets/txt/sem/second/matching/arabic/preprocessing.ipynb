```python
# File paths
corrections_file = "arabic_corrections_final.txt"
translations_file = "aligned_filtered_translations.txt"
output_file = "aligned_filtered_translations_corrected.txt"

# Step 1: Load corrections
# Assuming tab-separated: English[TAB]Corrected Arabic
corrections = {}
with open(corrections_file, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split("\t")
        if len(parts) >= 2:
            english, corrected_arabic = parts[0], parts[1]
            corrections[english] = corrected_arabic

# Step 2: Read original translations and apply corrections
updated_lines = []
with open(translations_file, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split("\t")
        if len(parts) >= 2:
            english, arabic = parts[0], parts[1]
            if english in corrections:
                arabic = corrections[english]  # replace with corrected
            updated_lines.append(f"{english}\t{arabic}\n")
        else:
            # keep line as-is if format is unexpected
            updated_lines.append(line)

# Step 3: Save corrected translations
with open(output_file, "w", encoding="utf-8") as f:
    f.writelines(updated_lines)

print(f"Corrections applied. Saved to '{output_file}'.")
```
